## Q1. What is the mathematical formula for a linear SVM?

The mathematical formula for a linear SVM is based on the concept of finding the hyperplane that best separates the two classes in the dataset. The hyperplane is defined by the equation:

$$w^Tx + b = 0$$

where `w` is the weight vector and `b` is the bias term. The decision boundary is given by:

$$w^Tx + b = 1$$

for the positive class and

$$w^Tx + b = -1$$

for the negative class. The distance between these two hyperplanes is called the margin, and the goal of SVM is to maximize this margin while minimizing the classification error.

The optimization problem for a linear SVM can be formulated as:

$$\min_{w,b} \frac{1}{2}||w||^2$$

subject to:

$$y_i(w^Tx_i + b) \geq 1, i = 1,2,...,n$$

where `y_i` is the class label of the `i-th` training example, `x_i` is its feature vector, and `n` is the number of training examples.

This optimization problem can be solved using quadratic programming (QP) techniques.

## Q3. What is the kernel trick in SVM?

The kernel trick is a powerful technique that enables SVMs to solve non-linear classification problems by implicitly mapping the input data to a higher-dimensional feature space. By doing so, it allows us to find a hyperplane that separates the different classes of data. The kernel function is used to compute the dot product between two vectors in the higher-dimensional feature space, without actually computing the coordinates of the vectors in that space. This makes it computationally efficient and allows us to work with infinite-dimensional feature spaces. 

The kernel function can be any function that satisfies the Mercer's condition, which ensures that the dot product between any two vectors in the feature space can be expressed as a function of their original coordinates in the input space. Some commonly used kernel functions include:
- Linear kernel
- Polynomial kernel
- Radial basis function (RBF) kernel
- Sigmoid kernel

The choice of kernel function depends on the nature of the data and the problem at hand.

## Q4. What is the role of support vectors in SVM Explain with example

Support vectors are the data points that are closest to the hyperplane and influence the position and orientation of the hyperplane. These data points are used to define the margin, which is the distance between the hyperplane and the closest data points from each class. The goal of SVM is to maximize this margin while minimizing the classification error.

In other words, support vectors are the critical data points that determine the position of the decision boundary. If we remove any of these data points, the position of the decision boundary would change. 

Here's an example to illustrate this concept. Suppose we have a dataset with two classes that are not linearly separable in 2D space. We can use a kernel function to map this dataset to a higher-dimensional feature space where it is linearly separable. In this new feature space, we can find a hyperplane that separates the two classes.

The support vectors are the data points that lie closest to this hyperplane. In other words, they are the data points that lie on or within the margin. These support vectors define the position and orientation of the hyperplane.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Generate some random data
X = np.random.randn(200, 2)
Y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)

# Fit SVM model
clf = svm.SVC(kernel='linear')
clf.fit(X, Y)

# Plot decision boundary
plt.scatter(X[:, 0], X[:, 1], c=Y, s=30)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

## Q6. SVM Implementation through Iris dataset.

### Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
### Train a linear SVM classifier on the training set and predict the labels for the testing setl
### Compute the accuracy of the model on the testing setl
### Plot the decision boundaries of the trained model using two of the featuresl
### Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pandas as pd

# Load the iris dataset
df = datasets.load_iris()
X = df.data[:, :2]  # Consider only the first two features
y = df.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pd.DataFrame(X)[0]
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
## prediction 
y_pred = svc.predict(X_test)
```

```python
svc.coef_
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Plot the decision boundaries of the trained model using two features
def plot_decision_boundary(model, X, y):
    h = 0.02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title('Decision Boundary of SVM')
    plt.show()

plot_decision_boundary(svc, X, y)
```

```python
from sklearn.model_selection import GridSearchCV
#defining the parameter ranges

para_grid = {'C':[0.1, 1, 10,100,1000],
             'gamma':[1,0.1,0.01, 0.001, 0.0001],
             'kernel':['linear']
            }
grid = GridSearchCV(SVC(),param_grid=para_grid, refit = True, cv = 5, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## prediction
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
